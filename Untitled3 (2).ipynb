## 과제6: 이진 검색 트리 문제 풀이

**1번 1)**

**답:** 7, 17, 70으로 3개이다.


**1번 2)**

**답:** 이진 트리에서 높이가 h일때 리프노드가 최대가 되려면 포화 이진 트리 형태가 되어야한다. 2^(4-1) = 8, 가능한 리프 노드와 최대 개수는 8개이다.

**1번 3)**

**답:** 높이가 4인 포화 이진 트리의 리프노트 수 : 8개, 내부 노드 수 : 7개. 1개 차이가 난다.

**1번 4)**

**답:** 위 3)의 성질은 높이가 h인 모든 포화 이진 트리에 성립한다.

---

**2번 1)**

**답:** 트리의 총 노드 수가 n이면 리프노드의 수는 (n/2)개 이다.

**2번 2)**

**답:** 리프 노트의 수 = 내부 노드의 수 + 1이다.

---

**3번 1)**

**답:** 전위 순회는 루트를 제일 먼저, 그리고 왼쪽, 오른쪽 순서로 방문하므로 5, 3, 1, 8, 7, 6, 9 이다.

**3번 2)**

**답:** 중위 순회는 루트를 중간에 방문하므로 1, 3, 5, 6, 7, 8, 9 이다.

**3번 3)**

**답:** 후위 순회는 루트를 맨 뒤에 방문하므로 1, 3, 6, 7, 9, 8, 5

**3번 4)**

**답:** 이진검색 트리 T의 중위 순회는 노드들을 정렬된 순서로 방문한다.

베이스케이스 : h=1. T는 단 한 개의 노드를 갖는다. 유일한 노드를 방문하는 것은 당연히 정렬된 순서다.

귀납적 가정: k < h인 모든 k에 대해 성립한다고 가정,
귀납적 전개: 귀납적 가정에 의해 중위 순회는 왼쪽 노드와 오른쪽 노드를 정렬된 순서로 방문함.

왼쪽 자식 노드의 모든 키는 그 키의 부모 트리의 키보다 작고 오른쪽 자식 노드는 부모 보드의 키보다 크므로 중위 순회는 정렬된 순서이다.
